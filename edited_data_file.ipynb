## 2) Process Data from Dirty to Clean

### 2.1) Data Transformation

```python
# Sử dụng dòng đầu tiên làm Header
df.columns=df.iloc[0]
df=df[1:]
df
```

```python
# Unpivot a dataframe
df_unpivoted = df.melt(id_vars=['visitor_id'], var_name='create_date', value_name='Number_of_Orders')
df_unpivoted
```

```python
# Hiển thị một số thông tin cơ bản
df_unpivoted.info()
```

```python
# Tạo thêm 1 cột 'create_date_new_format' để quy đổi cột 'create_date' giá trị từ 43770 đến 43869 sang 11/1/2019 đến 2/8/2020
dict={}
create_date_unique=df_unpivoted['create_date'].unique()
date=datetime.date(month=11,day=1,year=2019)
time_delta=datetime.timedelta(days=1)
for key in create_date_unique:
    dict[key]=str(date)
    date+=time_delta

create_date_unique=[]
for index in df_unpivoted.index:
    create_date_unique.append(dict[df_unpivoted['create_date'][index]])
df_unpivoted['create_date_new_format']=create_date_unique

df_unpivoted
```

## 2.2) Cleaning data

```python
# Xoá cột create_date (vì đã thay thế bằng cột 'create_date_new_format')
df_unpivoted=df_unpivoted.drop(labels='create_date',axis=1 )

# Đặt lại tên cho cột 'create_date_new_format' là 'create_date'
df_unpivoted=df_unpivoted.rename(columns={'create_date_new_format':'create_date'})
df_unpivoted
```

### 2.3) Save the file

```python
# Lưu lại file data vừa chỉnh sửa xong
df_unpivoted.to_csv(r"C:\Users\VLN\Documents\Project_BusinessCase\edited_data_file.csv",index=False)
```

## 2) Kiểm tra xem tất cả các id trong file data đều có trong file visitor_information hay không ?

```python
# Tạo list đặt tên là visitor_only chứa những visitor_id có trong file data nhưng không có trong file visitor_information
# lưu ý: 'visitor_only' là những khách hàng chỉ ghé nhưng chưa mua hàng & 'buyer' là những khách hàng đã mua hàng
visitor_only=[]
temp=data['visitor_id'].unique()
for element in visitor_info['visitor_id'].unique():
    if element not in temp:
        visitor_only.append(element)
```

```python
nO_visitor_only= len(visitor_only)
nO_buyer=len(temp)

print('Number of visitor_only: ' + str(nO_visitor_only))
print('Number of buyer: '+ str(nO_buyer))


total_visitorOnlyBuyer=nO_visitor_only+nO_buyer
nO_visitor_id_Info=len(visitor_info['visitor_id'].unique())

# So sánh để tìm xem có id nào không nằm trong file visitor_information không
print('Number of visitor_only + buyer: '+ str(total_visitorOnlyBuyer))
print('Number of visitor_id in file visitor_information: '+str(nO_visitor_id_Info))
delta=total_visitorOnlyBuyer-nO_visitor_id_Info
print('Difference between the two numbers above: '+str(delta))
```

```python
# Sau khi so sánh thì có 1 id bị lệch và id đó nằm ở trong file data
# Kiểm tra xem id bị lệch đó là id nào ? 
a=[]
for element in data['visitor_id'].unique():
    if element not in visitor_info['visitor_id'].unique():
        a.append(element)
a
```
